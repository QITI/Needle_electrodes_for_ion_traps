```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('needlepictures/pic3.jpg')
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray,cmap='gray')
```

```python
image_gray = cv2.medianBlur(image_gray, 1)
plt.imshow(image_gray,cmap='gray')
```

```python
edges = cv2.Canny(image_gray,threshold1=30, threshold2=150)
```

```python
plt.imshow(edges,cmap='gray')
```
