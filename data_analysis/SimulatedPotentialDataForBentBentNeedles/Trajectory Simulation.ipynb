```python
import IonTracs as it
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# read the potential data created by COMSOL
dataRF = pd.read_csv('RFpotential.csv',skiprows=8)
dataDC = pd.read_csv('DCpotential.csv',skiprows=8)
RFdata = np.array(dataRF.values)
DCdata = np.array(dataDC.values)
del dataRF
del dataDC

# Iontracs has default as microns for x,y,z data  
x = np.unique(RFdata[:,0])*1e6
y = np.unique(RFdata[:,1])*1e6
z = np.unique(RFdata[:,2])*1e6

# create a data structure that Iontracs can understand
def GetBasisPotentials(RFdata, DCdata, config=3):
    Ubasis = np.zeros([x.size, y.size, z.size, 2])
    # first three dimensions of Ubasis store potential
    # last dimentions: 0 is RF and 1 is DC
    Ubasis[:, :, :, 0] = RFdata[:, config].reshape(x.size, y.size, z.size, order='F')
    Ubasis[:, :, :, 1] = DCdata[:, config].reshape(x.size, y.size, z.size, order='F')
    return Ubasis # 4 dimensional structure with first 3 dimensions for storing potential ans last for electrode number 
```

```python
# potentials for the bent needles
Ubasis_asym = GetBasisPotentials(RFdata,DCdata,config=11) 

# potential for the symmetric needles
Ubasis_sym = GetBasisPotentials(RFdata,DCdata,config=3) 
```

```python
#define trajectory object
traj_sym = it.trajectory(ubasis=Ubasis_sym,x=x,y=y,z=z,vrf_amp=250,vrf_freq=20,\
                     rfelec_in_phase=[1,0],rfelec_out_phase=[0,0],dcelec=[0,50])
```

```python
# Solve for the trajectory
traj_sym.GetTrajectory(r_0=[0,0,0,0,0,0],t_values=np.linspace(0,15,100000),rotate_xy=True)
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111)
ax.plot(traj_sym.x_traj[:50000],traj_sym.y_traj[:50000])
#plt.savefig('trajectory.svg')
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(traj_sym.x_traj[:50000],traj_sym.z_traj[:50000])
plt.savefig('trajectoryXZ.svg')
plt.show()
```

```python
# Frequency spectrum is only calculated after running the following command
traj_sym.GetFrequencySpectrum(noreturn=True)
```

```python
plt.plot(traj_sym.freq_axis[10:400],traj_sym.ampl_x[10:400])
```

```python
freq_axis = my_traj_0deg.freq_axis
fft_sym = my_traj_0deg.ampl_x
fft_asym1 = my_traj_0p5deg.ampl_x
fft_asym2 = my_traj_1deg.ampl_x
fft_asym3 = my_traj_2deg.ampl_x
```

```python
bool_mask = (freq_axis>0.5) * (freq_axis<22)
freq_axis = freq_axis[bool_mask]
fft_sym = fft_sym[bool_mask]
fft_asym1 = fft_asym1[bool_mask]
fft_asym2 = fft_asym2[bool_mask]
fft_asym3 = fft_asym3[bool_mask]
fft_sym = fft_sym/np.max(fft_sym)
fft_asym1 = fft_asym1/np.max(fft_asym1)
fft_asym2 = fft_asym2/np.max(fft_asym2)
fft_asym3 = fft_asym3/np.max(fft_asym3)
```

```python
plt.plot(freq_axis,fft_sym)
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot()
ax.plot(freq_axis/20,fft_asym1,'b')
ax.plot(freq_axis/20,fft_asym2,'r')
#ax.plot(freq_axis,fft_asym3,label='bent needle3')
ax.plot(freq_axis/20,fft_sym,'g')
ax.tick_params(axis='x', labelsize= 18)
ax.tick_params(axis='y', labelsize= 22)
ax.set_xlabel('Normalized Frequency',fontsize = 18)
ax.set_ylabel('Normalized amplitude',fontsize = 18)
#ax.legend(fontsize=15)
plt.savefig('fullview.svg')
plt.show()
```
